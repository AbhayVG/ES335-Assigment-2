# Task 5 : Image Compression using Matrix Factorisation 

You have an image patch of size (50x50) that you want to compress using matrix factorization. To do this, you'll split the patch $[N\times N]$ into two smaller matrices of size $[N\times r]$ and $[r\times N]$ using matrix factorization. Compute the compressed patch by multiplying these two matrices and compare the reconstructed image patch with the original patch. Compute the Root Mean Squared Error (RMSE) and Peak Signal-to-Noise Ratio (PSNR) between the original and reconstructed image patches. 

- Test different values for the low-rank $r = [5, 10, 25, 50]$.
- Use Gradient Descent to learn the compressed matrices.
- Display the reconstructed image patches, keeping the original pixel values outside the patch unchanged, and use your compressed matrix for the patch to show how well the reconstruction works.
 - Compute the RMSE and PSNR for each value of $r$. 


```python
#  load the original image which is 1.jpg and extract a patch of size 50 x 50.
#  Display the original image and the extracted patch.
import cv2
import numpy as np
import matplotlib.pyplot as plt
# Load the original image
img = cv2.imread('1.jpg')
# Extract a patch of size 50 x 50
patch = img[90:140, 90:140]
# Now denote the extracted patch as matrix X
X = patch
# Display the extracted patch
plt.figure()
plt.imshow(cv2.cvtColor(patch, cv2.COLOR_BGR2RGB))
plt.title('Extracted Patch')
plt.axis('off')
plt.show()

# save the extracted patch
cv2.imwrite('patch2.jpg', patch)

```

```python
def matrix_factorization(X, r, num_iterations=15000, learning_rate=0.00001):
    N = X.shape[0]
    M = X.shape[1]
    W = np.random.rand(N, r)
    H = np.random.rand(r, M)
    for i in range(num_iterations):
        X_hat = np.dot(W, H)
        loss = np.linalg.norm(X - X_hat)
        if i % 100 == 0:
            print(f'Iteration: {i}, Loss: {loss}')
        W_grad = np.dot((X_hat - X), H.T)
        H_grad = np.dot(W.T, (X_hat - X))
        W = W - learning_rate * W_grad
        H = H - learning_rate * H_grad
    return W, H
    
```

```python
# Load the image patch
patch = cv2.imread('patch2.jpg')
# Reshape the patch to be a 2D matrix
X = patch.reshape(50, 50*3)
# Test different values for the low-rank r
r_values = [5, 10, 25, 50]
for r in r_values:
    # Perform matrix factorization
    W, H = matrix_factorization(X, r)
    # Compute the compressed patch
    X_hat = np.dot(W, H)
    # Reshape the compressed patch to be a 3D matrix
    patch_hat = X_hat.reshape(50, 50, 3)
    
   # Display the rankwise patch reconstruction with iteration and loss
    plt.figure()
    plt.imshow(cv2.cvtColor(patch_hat.astype(np.uint8), cv2.COLOR_BGR2RGB))
    plt.title(f'Rank: {r}')
    plt.axis('off')
    plt.show()
    cv2.imwrite(f'patch2_rank_{r}.jpg', patch_hat.astype(np.uint8))
    
    
```

```python
# Now print only the rmse and psnr for each value of r. no need to print all iteration and loss. consider those things from previous cell. No need to write whole above code just call things from there.
def matrix_factorization(X, r, num_iterations=15000, learning_rate=0.00001):
    N = X.shape[0]
    M = X.shape[1]
    W = np.random.rand(N, r)
    H = np.random.rand(r, M)
    for i in range(num_iterations):
        X_hat = np.dot(W, H)
        loss = np.linalg.norm(X - X_hat)
        if i % 100 == 0:
            print(f'Iteration: {i}, Loss: {loss}')
        W_grad = np.dot((X_hat - X), H.T)
        H_grad = np.dot(W.T, (X_hat - X))
        W = W - learning_rate * W_grad
        H = H - learning_rate * H_grad
    return W, H

def rmse_psnr(X, X_hat):
    rmse = np.sqrt(np.mean((X - X_hat) ** 2))
    psnr = 20 * np.log10(255 / rmse)
    return rmse,psnr
# call the function from previous cell and print only rmse and psnr for each value of r
# Load the image patch
patch = cv2.imread('patch1.jpg')
# Reshape the patch to be a 2D matrix
X = patch.reshape(50, 50*3)
# Test different values for the low-rank r
r_values = [5, 10, 25, 50]

for r in r_values:
    # Perform matrix factorization
    W, H = matrix_factorization(X, r)
    # Compute the compressed patch
    X_hat = np.dot(W, H)
    # Reshape the compressed patch to be a 3D matrix
    patch_hat = X_hat.reshape(50, 50, 3)
    rmse, psnr = rmse_psnr(X, X_hat)
    print(f'Rank: {r}, RMSE: {rmse}, PSNR: {psnr}')

```

```python
# add the reconstructed patch in the original image and show a bounding box there in red.
# Load the original image
img = cv2.imread('1.jpg')
# Load the reconstructed patch
patch_hat = cv2.imread('patch2_rank_50.jpg')
# Add the reconstructed patch to the original image
img[90:140, 90:140] = patch_hat
# Draw a bounding box in red
cv2.rectangle(img, (90, 90), (140, 140), (0, 0, 255), 2)
# Display the original image with the reconstructed patch
plt.figure()
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image with Reconstructed Patch With r = 50')
plt.axis('off')
plt.show()
# save the image
cv2.imwrite('original_with_patch2.jpg', img)

```
